```python
# Kernel Density Estimate plot using random dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn
%matplotlib inline
```

```python
plotData = pd.DataFrame(randn(400,3), columns = ['Red', 'Blue', 'Green'])  # randn(rows,columns)
plotData
```

```python
plotData.plot.kde()          # kde from scipy
plt.show()
```
