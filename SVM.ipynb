```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
dir(iris)                
```

```python
iris.feature_names
iris.target           
```

```python
petal_data = pd.DataFrame(iris.data, columns = iris.feature_names)
petal_data
```

```python
petal_data['iris_target'] = iris.target
petal_data
```

```python
petal_data['target name'] = petal_data.iris_target.apply(lambda x: iris.target_names[x])
petal_data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
setosaD = petal_data[petal_data.iris_target == 0]          # assign each iris type (0 , 1, 2) to variables
versicolD = petal_data[petal_data.iris_target == 1]
virginicaD = petal_data[petal_data.iris_target == 2]
```

```python
setosaD.head()
```

```python
# visualization using scatter plot
plt.xlabel('sepal length(cm)')
plt.ylabel('sepal width(cm)')
plt.scatter(setosaD['sepal length (cm)'], setosaD['sepal width (cm)'], color = 'red', marker= '*')
plt.scatter(versicolD['sepal length (cm)'], versicolD['sepal width (cm)'], color = 'green', marker= '.')
plt.scatter(virginicaD['sepal length (cm)'], virginicaD['sepal width (cm)'], color = 'blue', marker= '+')
plt.show()
```

```python
plt.xlabel('petal length(cm)')
plt.ylabel('petal width(cm)')
plt.scatter(setosaD['petal length (cm)'], setosaD['petal width (cm)'], color = 'red', marker= '*')
plt.scatter(versicolD['petal length (cm)'], versicolD['petal width (cm)'], color = 'green', marker= '.')
plt.scatter(virginicaD['petal length (cm)'], virginicaD['petal width (cm)'], color = 'blue', marker= '+')
plt.show()
```
